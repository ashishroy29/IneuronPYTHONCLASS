# While LOOP

```python
# it can be a statement which can be used for parsing a dataset or creating a loops
# While statement will always try to check, wether it is going to return true or false,
# if it is going to return true only in that case it will execute the loop otherwise it is going to break the loop
```

```python
a = 1
while a < 6 :
    print(a)
    a = a +1
```

```python
# for loop is going to work well in terms of collections and while loop is going to work well in terms of conditions
```

```python
a = 1
while a < 5 :
    print(a)
    if a == 4:
        break
    a = a + 1
```

```python
# Continue will help, to continue directly to the while loop with out executing the next one.
a = 1
while a < 5 :
    print(a)
    if a == 3:
        continue
    a = a + 1
```

```python
# where to use a break statement to stop the infinte loop
a = 1
while a < 5 :
    print(a)
    a = a + 1
    if a == 3:
        continue
```

```python
# pass statement, we just want to write a deadcode, it is not going to give me a syntatical error
while a < 5 :
    pass
```
